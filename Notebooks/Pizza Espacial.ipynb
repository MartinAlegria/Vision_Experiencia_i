```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Read Images

```python
#Read images from disk
image= cv2.imread('images/pizza_bluescreen.jpg')
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#Read space bg
space = cv2.imread('images/space_background.jpg')
space_rgb = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.imshow(space_rgb)
```

### Define thresholds for blue

```python
low_blue = np.array( [0,0,200])
upper_blue = np.array( [80,80,255])
```

```python
mask = cv2.inRange(img_rgb, low_blue, upper_blue)
plt.imshow(mask, cmap = 'gray')
```

### Mask the image

```python
masked_img = np.copy(img_rgb)
masked_img[mask!=0] = [0,0,0]

plt.imshow(masked_img)
```
